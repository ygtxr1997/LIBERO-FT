```python
import os
import re
import glob
from tqdm import tqdm
from libero.force.hdf5_replayer import HDF5Replayer
from libero.force.hdf5_compare import compare_hdf5
```

## Start to replay

```python
dataset_root = "/home/geyuan/code/LIBERO-FT/libero/datasets/"
dataset_subname = "libero_90"
output_dir = os.path.join(dataset_root, "libero_force")
os.makedirs(output_dir, exist_ok=True)

# pat = re.compile(r"(open|close).*drawer|microwave", re.IGNORECASE)  # 按需改 regex
pat = re.compile(r"(STUDY_SCENE3).*left", re.IGNORECASE)  # 按需改 regex

input_hdf5_fns = sorted(glob.glob(dataset_root + dataset_subname + "/*.hdf5"))
input_hdf5_fns = [fn for fn in input_hdf5_fns if pat.search(os.path.basename(fn))]
# print("Found hdf5 files:", input_hdf5_fns)
output_hdf5_fns = [
    os.path.join(
        output_dir,
        os.path.basename(fn).replace(".hdf5", "_wrench.hdf5")
    )
    for fn in input_hdf5_fns
]
print("Output hdf5 files:", output_hdf5_fns)


total_cnt = len(input_hdf5_fns)
for idx in tqdm(range(total_cnt), desc="Replaying"):
    input_hdf5_path = input_hdf5_fns[idx]
    output_hdf5_path = output_hdf5_fns[idx]

    replayer = HDF5Replayer(
        input_hdf5_path,
    )
    replayer.init_env()

    replayer.replay_hdf5_by_actions(
        hdf5_save_path=output_hdf5_path,
        video_save_dir="/home/geyuan/code/LIBERO-FT/debug/replay_videos/",
    )
    replayer.close_env()
```

```python
hdf5_fn = "KITCHEN_SCENE4_close_the_bottom_drawer_of_the_cabinet_and_open_the_top_drawer_demo.hdf5"
hdf5_path = os.path.join(dataset_root, dataset_subname, hdf5_fn)
hdf5_save_path = "tmp_replayed_wrench.hdf5"

replayer = HDF5Replayer(
    hdf5_path,
    # benchmark_task_suite_name="libero_90",
)
# replayer.vis_hdf5_contents()

replayer.close_env()
replayer.init_env()

replayer.replay_hdf5_by_actions(
    hdf5_save_path=hdf5_save_path,
)
```
